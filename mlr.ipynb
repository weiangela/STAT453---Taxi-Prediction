```python
import pandas as pd
import pyarrow.parquet as pq
import numpy as np
```

```python
data = pq.read_table('/content/drive/MyDrive/cleaned_taxi_data.parquet')
data = data.to_pandas()

data.describe()
```

```python
data.columns
# tip_bin is our target variable
# need to drop tip_pct, tip_amount
```

```python
X = data.drop(['tip_pct', 'tip_amount', 'tip_bin'], axis=1)
y = data['tip_bin']

print('Class labels:', np.unique(y))
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y)
from sklearn.preprocessing import StandardScaler

#normalize data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
import time
#if we want to we can do multiple variations of logistic regressions ie we can add penalities or change the solver
t0=time.time()
model = LogisticRegression(multi_class='multinomial', max_iter = 2000)
model.fit(X_train, y_train)
print("Training time:", time.time()-t0)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#using accuracy confusion matrix to evaluate model
#again can add other metrics if we want

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

confusionmatrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{confusionmatrix}")
```
